```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_stats(data, type, xlabel, ylabel, title):
    if type=='hist':
        sns.histplot(data)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.title(title)
    if type=='scatter':
        sns.scatterplot(data)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.title(title)
    return()
```

```python
data_df=pd.DataFrame([np.random.rand(100), np.random.rand(100), np.random.rand(100)], index=['Random 1', 'Random 2', 'Random 3']).T
data_df
```

```python
plot_stats(data_df,'scatter', 'Index', 'Random Value', 'Scatterplot')
```

```python
plot_stats(data_df,'hist', ' Random Value', 'Count', 'Histogram')
```
